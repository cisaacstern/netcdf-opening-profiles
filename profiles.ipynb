```python
import gcsfs
import xarray as xr

fs_gcs = gcsfs.GCSFileSystem(anon=True)
cache_base = "pangeo-forge-us-central1/pangeo-forge-cache"

def open_from_cache(recipe_name):
    path = fs_gcs.ls(f"{cache_base}/{recipe_name}")[0]
    print(
        f"Filename: {path.split('/')[-1]}\n"
        f"On disk size: {fs_gcs.du(path)/1e9:.02f} GB."
    )
    return fs_gcs.open(path)
```

```python
enatl60_example_file = open_from_cache("swot_adac/eNATL60/Region01/interior_daily/aso")
```

```python
%%prun -D enatl60-region01-interior-aso-2009-09.prof
enatl60_ds = xr.open_dataset(fs_gcs.open(enatl60_example_file))
```
